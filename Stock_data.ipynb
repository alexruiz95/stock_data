```python
import yfinance as yf

#define the ticker symbol
tickerSymbol = 'MSFT'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices for this ticker
tickerDf = tickerData.history(period='1d', start='2020-1-1', end='2020-5-25')

#period: the frequency at which to gather the data; common options would include ‘1d’ (daily), ‘1mo’ (monthly), ‘1y’ (yearly)
#start: the date to start gathering the data. For example ‘2010–1–1’
#end: the date to end gathering the data. For example ‘2020–1–25’

#see your data
tickerDf
```

```python
# lets make a simple conditional that checks to see if the closing is greater then the open 
df=tickerDf

df.loc[df['Open'] < df['Close'], 'Open less then close'] = 'True'+int(df.Open-df.Close)
df.loc[df['Open'] > df['Close'], 'Open less then close'] = 'False'
# df.loc[df['set_of_numbers'] > 4, 'equal_or_lower_than_4?'] = 'False' 

print (df['Open less then close'])
```
